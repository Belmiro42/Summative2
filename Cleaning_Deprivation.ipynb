```python
import pandas as pd
import matplotlib.pyplot as plt
```

Open Files



```python
files = ["./Data/Deprivation/Deprivation_2015.xlsx", "./Data/Deprivation/Deprivation_2019.xlsx", "./Data/Deprivation/Deprivation_2025.xlsx"]
dfs = [pd.read_excel(f, sheet_name=1) for f in files]
```

```python
dfs[0]
```

```python
for i in range(len(files)):
    dfs[i] = dfs[i].iloc[:, [2, 4]]
```

```python
for i in range(len(files)):
    dfs[i] = dfs[i].groupby(dfs[i].iloc[:, 0]).mean(numeric_only=True)
```

```python
dates = ["2015 Average IMD Score", "2019 Average IMD Score", "2025 Average IMD Score"] 
for i in range(len(files)):
    dfs[i].index.name = "Council Code"
    dfs[i] = dfs[i].rename(columns={'Index of Multiple Deprivation (IMD) Score': dates[i]})
```

```python
for i in range(len(files)):
    if (i==0):
        dfs_merged = dfs[0]
    else:
        first_left = dfs_merged.columns[0]
        first_right = dfs[i].columns[0]
        dfs_merged = dfs_merged.merge(dfs[i], left_index=True, right_index=True)
```

```python
dfs_merged.to_csv("./CleanedData/DeprivationCleaned.csv", index=True)
```
